### **Serie de Fourier (Formas Exponencial, Trigonométrica y Compacta)**

* **Aplicación**: Se utiliza para analizar señales periódicas en el dominio del tiempo continuo.
* **Espectro**: Discreto, compuesto por frecuencias que son múltiplos enteros de la frecuencia fundamental.
* **Semejanzas**: Todas las variantes (exponencial, trigonométrica y compacta) representan la misma señal periódica, solo cambian en su forma de expresión.
* **Diferencias**:

  * La forma **trigonométrica** emplea funciones seno y coseno.
  * La forma **exponencial** usa exponenciales complejas.
  * La forma **compacta** describe la señal usando magnitud y fase.

---

### **Transformada de Fourier**

* **Aplicación**: Se aplica a señales no periódicas en tiempo continuo.
* **Espectro**: Continuo, abarca todas las frecuencias reales.
* **Semejanzas**: Se considera una generalización de la serie de Fourier cuando el periodo tiende a infinito.
* **Diferencias**: A diferencia de la serie de Fourier, trabaja con señales no periódicas y genera un espectro continuo.

---

### **Transformada de Fourier en Tiempo Discreto (DTFT)**

* **Aplicación**: Diseñada para señales no periódicas en tiempo discreto.
* **Espectro**: Continuo y periódico en la frecuencia, con un periodo de $2\pi$ radianes.
* **Semejanzas**: Es el equivalente en tiempo discreto de la transformada de Fourier.
* **Diferencias**: A diferencia de la versión continua, la DTFT trabaja con secuencias discretas y su espectro se repite periódicamente.

---

### **Transformada Discreta de Fourier (DFT)**

* **Aplicación**: Se utiliza con secuencias finitas de tiempo discreto, asumiendo periodicidad.
* **Espectro**: Discreto y finito, se evalúa solo en ciertas frecuencias.
* **Semejanzas**: Representa una versión muestreada de la DTFT y es esencial en aplicaciones digitales.
* **Diferencias**: A diferencia de la DTFT, se limita a un conjunto finito de muestras tanto en el tiempo como en la frecuencia.

---

### **Algoritmo de Transformada Rápida de Fourier (FFT)**

* **Definición**: No es una nueva transformada, sino un algoritmo optimizado para calcular la DFT de forma eficiente.
* **Importancia**: Reduce drásticamente el tiempo de cálculo, lo que ha sido clave en áreas como compresión, análisis de espectro y filtrado digital.
* **Funcionamiento**:

  * El algoritmo FFT más conocido (Cooley-Tukey) usa una estrategia de "divide y vencerás" para descomponer una DFT grande en subproblemas más pequeños.
  * Se aprovechan propiedades de simetría y periodicidad de los exponenciales complejos.
  * Esta descomposición continúa hasta obtener DFTs de tamaño 1, que luego se combinan para formar el resultado final.
  * Existen variantes del algoritmo como radix-2 y radix-4, según cómo se realice la descomposición.

---

### **Costo Computacional**

* **DFT directa**: Requiere aproximadamente $O(N^2)$ operaciones complejas.
* **FFT**: Reduce ese costo a $O(N \log_2 N)$ si $N$ es una potencia de 2.
* **Comparación**: Por ejemplo, para $N = 1024$:

  * La DFT directa necesita alrededor de 1,000,000 de operaciones.
  * La FFT solo requiere unas 10,240 operaciones.
  * Esto representa una mejora sustancial en eficiencia, especialmente importante en aplicaciones en tiempo real.

Donde:

$X(\omega)$ es la transformada de Fourier de $x(t)$.
$\omega$ es la frecuencia angular en radianes por segundo.
$j$ es la unidad imaginaria ($\sqrt{-1}$).
Aquí te presento cómo abordar cada inciso, aplicando la definición de la transformada de Fourier:

a) $x(t) = e^{-a|t|}, \quad a \in \mathbb{R}^{+}$

Necesitamos evaluar las dos integrales que establecimos: $$ X(\omega) = \int_{-\infty}^{0} e^{(a-j\omega)t} dt + \int_{0}^{\infty} e^{-(a+j\omega)t} dt $$

Para la primera integral: $$ \int_{-\infty}^{0} e^{(a-j\omega)t} dt = \left[ \frac{e^{(a-j\omega)t}}{a-j\omega} \right]_{-\infty}^{0} $$ Dado que $a > 0$, cuando $t \to -\infty$,  $e^{(a-j\omega)t} \to 0$. $$ = \frac{e^0}{a-j\omega} - 0 = \frac{1}{a-j\omega} $$

Para la segunda integral: $$ \int_{0}^{\infty} e^{-(a+j\omega)t} dt = \left[ \frac{e^{-(a+j\omega)t}}{-(a+j\omega)} \right]_{0}^{\infty} $$ Dado que $a > 0$, cuando $t \to \infty$, $e^{-(a+j\omega)t} \to 0$. $$ = 0 - \frac{e^0}{-(a+j\omega)} = \frac{1}{a+j\omega} $$

Sumando los dos resultados: $$ X(\omega) = \frac{1}{a-j\omega} + \frac{1}{a+j\omega} = \frac{a+j\omega + a-j\omega}{(a-j\omega)(a+j\omega)} = \frac{2a}{a^2 + \omega^2} $$

Entonces, la Transformada de Fourier de $e^{-a|t|}$ es $X(\omega) = \frac{2a}{a^2 + \omega^2}$.

b) $x(t) = \cos(w_c t), \quad w_c \in \mathbb{R}$

Tenemos: $$ X(\omega) = \frac{1}{2} \int_{-\infty}^{\infty} e^{j(w_c - \omega)t} dt + \frac{1}{2} \int_{-\infty}^{\infty} e^{-j(w_c + \omega)t} dt $$ Sabemos que $\int_{-\infty}^{\infty} e^{j\alpha t} dt = 2\pi \delta(\alpha)$

Entonces, la primera integral es: $\int_{-\infty}^{\infty} e^{j(w_c - \omega)t} dt = 2\pi \delta(w_c - \omega) = 2\pi \delta(\omega - w_c)$.

Y la segunda integral es $\int_{-\infty}^{\infty} e^{-j(w_c + \omega)t} dt = \int_{-\infty}^{\infty} e^{j(-(w_c + \omega))t} dt = 2\pi \delta(-(w_c + \omega)) = 2\pi \delta(w_c + \omega) = 2\pi \delta(\omega + w_c)$.

Sustituyendo de nuevo: $$ X(\omega) = \frac{1}{2} (2\pi \delta(\omega - w_c)) + \frac{1}{2} (2\pi \delta(\omega + w_c)) = \pi \delta(\omega - w_c) + \pi \delta(\omega + w_c) $$ Entonces, la Transformada de Fourier de $\cos(w_c t)$ es $X(\omega) = \pi (\delta(\omega - w_c) + \delta(\omega + w_c))$.

c) $x(t) = \sin(w_s t), \quad w_s \in \mathbb{R}$

La segunda integral es $\int_{-\infty}^{\infty} e^{-j(w_s + \omega)t} dt = 2\pi \delta(\omega + w_s)$.

Sustituyendo de nuevo: $$ X(\omega) = \frac{1}{2j} (2\pi \delta(\omega - w_s)) - \frac{1}{2j} (2\pi \delta(\omega + w_s)) = \frac{\pi}{j} \delta(\omega - w_s) - \frac{\pi}{j} \delta(\omega + w_s) $$ Dado que $\frac{1}{j} = -j$: $$ X(\omega) = -j\pi \delta(\omega - w_s) + j\pi \delta(\omega + w_s) = j\pi (\delta(\omega + w_s) - \delta(\omega - w_s)) $$ Entonces, la Transformada de Fourier de $\sin(w_s t)$$\sin(w_s t)$ es $X(\omega) = j\pi (\delta(\omega + w_s) - \delta(\omega - w_s))$.

d) $x(t) = f(t)\cos(w_c t), \quad w_c \in \mathbb{R}, \quad f(t) \in \mathbb{R} \text{ o } \mathbb{C}$

e) $x(t) = e^{-a|t|^2}, \quad a \in \mathbb{R}^{+}$

Necesitamos evaluar: $$ X(\omega) = \int_{-\infty}^{\infty} e^{-at^2} e^{-j\omega t} dt = \int_{-\infty}^{\infty} e^{-at^2 - j\omega t} dt $$ Para resolver esto usando la fórmula de la integral Gaussiana, completamos el cuadrado en el exponente: $$ -at^2 - j\omega t = -a \left( t^2 + \frac{j\omega}{a} t \right) $$ Para completar el cuadrado de $t^2 + \frac{j\omega}{a} t$, sumamos y restamos $(\frac{j\omega}{2a})^2$:
$$ t^2 + \frac{j\omega}{a} t + \left(\frac{j\omega}{2a}\right)^2 - \left(\frac{j\omega}{2a}\right)^2 = \left( t + \frac{j\omega}{2a} \right)^2 - \frac{(j\omega)^2}{4a^2} = \left( t + \frac{j\omega}{2a} \right)^2 + \frac{\omega^2}{4a^2} $$ Entonces el exponente se convierte en: $$ -a \left[ \left( t + \frac{j\omega}{2a} \right)^2 + \frac{\omega^2}{4a^2} \right] = -a \left( t + \frac{j\omega}{2a} \right)^2 - \frac{a\omega^2}{4a^2} = -a \left( t + \frac{j\omega}{2a} \right)^2 - \frac{\omega^2}{4a} $$ Ahora la integral es: $$ X(\omega) = \int_{-\infty}^{\infty} e^{-a \left( t + \frac{j\omega}{2a} \right)^2 - \frac{\omega^2}{4a}} dt = e^{-\frac{\omega^2}{4a}} \int_{-\infty}^{\infty} e^{-a \left( t + \frac{j\omega}{2a} \right)^2} dt $$ Sea $\tau = t + \frac{j\omega}{2a}$. En este caso, $x(t) = 3$, y tenemos $t^3$. La Transformada de Fourier de una constante $C$ es $2\pi C \delta(\omega)$. Por lo tanto, $\mathcal{F}\{3\} = 2\pi (3) \delta(\omega) = 6\pi \delta(\omega)$. Llamemos a esto $X(\omega)$.

Ahora aplicamos la propiedad de diferenciación con $n=3$: $\mathcal{F}\{t^3 \cdot 3\} = (j)^3 \frac{d^3}{d\omega^3} (6\pi \delta(\omega))$ $\mathcal{F}\{3t^3\} = -j \frac{d^3}{d\omega^3} (6\pi \delta(\omega))$ $\mathcal{F}\{3t^3\} = -6\pi j \frac{d^3}{d\omega^3} \delta(\omega)$

Las derivadas de la función delta de Dirac se definen en el sentido de las distribuciones. La tercera derivada de $\delta(\omega)$ se denota por $\delta'''(\omega)$.

Por lo tanto, la Transformada de Fourier de $3t^3$ es $-6\pi j \delta'''(\omega)$.

e) $\frac{B}{T} \sum_{n=-\infty}^{+\infty} \left( \frac{1}{a^2 + (w - n\omega_o)^2} + \frac{1}{a + j(w - n\omega_o)} \right)$,

donde:$$ ( n \in \{0, \pm1, \pm2, \dots\}, \quad \omega_o = \frac{2\pi}{T}, \quad B, T \in \mathbb{R}^+ )$$

Esta expresión parece la representación de la Serie de Fourier de una señal periódica, específicamente en el dominio de la frecuencia. El término $\sum_{n=-\infty}^{+\infty} F(\omega - n\omega_o)$ representa la Transformada de Fourier de una señal periódica formada al repetir una señal básica.

Consideremos los términos dentro de la suma:

Término 1: $\frac{1}{a^2 + (w - n\omega_o)^2}$. Este es el desplazamiento de $\frac{1}{a^2 + \omega^2}$. Sabemos que $\mathcal{F}\{e^{-a|t|}\} = \frac{2a}{a^2 + \omega^2}$. Por lo tanto, $\frac{1}{a^2 + \omega^2} = \frac{1}{2a} \mathcal{F}\{e^{-a|t|}\}$. El término desplazado $\frac{1}{a^2 + (w - n\omega_o)^2}$ es la Transformada de Fourier de $\frac{1}{2a} e^{-a|t|} e^{jn\omega_o t}$.

Término 2: $\frac{1}{a + j(w - n\omega_o)}$. Este es el desplazamiento de $\frac{1}{a + j\omega}$$\frac{1}{a + j\omega}$. Sabemos que $\mathcal{F}\{e^{-at} u(t)\} = \frac{1}{a + j\omega}$. El término desplazado $\frac{1}{a + j(w - n\omega_o)}$ es la Transformada de Fourier de $e^{-at} u(t) e^{jn\omega_o t}$.

La suma en el dominio de la frecuencia corresponde a una señal periódica en el dominio del tiempo. Si $X(\omega)$ es la Transformada de Fourier de $x(t)$, entonces la Transformada de Fourier de una señal periódica $x_p(t)$ con período $T$ y coeficientes de la serie de Fourier $c_n$ está dada por: $X_p(\omega) = \frac{2\pi}{T} \sum_{n=-\infty}^{+\infty} c_n \delta(\omega - n\omega_o)$, donde $\omega_o = \frac{2\pi}{T}$ y $c_n = \frac{1}{T} \int_0^T x_p(t) e^{-jn\omega_o t} dt$.

Alternativamente, si una señal $x(t)$ tiene una Transformada de Fourier $X(\omega)$, entonces una señal periódica formada al repetir $x(t)$ con período $T$ tiene una Transformada de Fourier: $\frac{2\pi}{T} \sum_{n=-\infty}^{\infty} X(n\omega_o) \delta(\omega - n\omega_o)$.

La expresión dada no está directamente en la forma de una suma de funciones delta, sino más bien una suma de versiones desplazadas de funciones de $\omega$. Esto sugiere que la señal original en el dominio del tiempo es una suma de términos asociados con cada función desplazada.

Sea $F(\omega) = \frac{1}{a^2 + \omega^2} + \frac{1}{a + j\omega}$. La expresión dada es $\frac{B}{T} \sum_{n=-\infty}^{+\infty} F(\omega - n\omega_o)$.

La Transformada Inversa de Fourier de la suma es la suma de las transformadas inversas: $\mathcal{F}^{-1}\left\{ \sum_{n=-\infty}^{+\infty} F(\omega - n\omega_o) \right\} = \sum_{n=-\infty}^{+\infty} \mathcal{F}^{-1}\{F(\omega - n\omega_o)\}$ $= \sum_{n=-\infty}^{+\infty} e^{jn\omega_o t} f(t)$, donde $f(t) = \mathcal{F}^{-1}\{F(\omega)\} = \frac{1}{2a} e^{-a|t|} + e^{-at} u(t)$. $= f(t) \sum_{n=-\infty}^{+\infty} e^{jn\omega_o t}$

La suma $\sum_{n=-\infty}^{+\infty} e^{jn\omega_o t}$ es la representación en serie de Fourier de un tren de impulsos delta de Dirac en el dominio del tiempo: $\frac{2\pi}{T} \sum_{k=-\infty}^{\infty} \delta(t - kT)$.

Entonces, la Transformada Inversa de Fourier de $\sum_{n=-\infty}^{+\infty} F(\omega - n\omega_o)$ es: $f(t) \left( \frac{2\pi}{T} \sum_{k=-\infty}^{\infty} \delta(t - kT) \right) = \frac{2\pi}{T} \sum_{k=-\infty}^{\infty} f(t) \delta(t - kT)$ Usando la propiedad $f(t) \delta(t - t_0) = f(t_0) \delta(t - t_0)$: $= \frac{2\pi}{T} \sum_{k=-\infty}^{\infty} f(kT) \delta(t - kT)$

La Transformada Inversa de Fourier de la expresión completa es multiplicada por $\frac{B}{T}$: $x(t) = \frac{B}{T} \left( \frac{2\pi}{T} \sum_{k=-\infty}^{\infty} f(kT) \delta(t - kT) \right)$ $x(t) = \frac{2\pi B}{T^2} \sum_{k=-\infty}^{\infty} f(kT) \delta(t - kT)$

Sustituimos $f(kT) = \frac{1}{2a} e^{-a|kT|} + e^{-akT} u(kT)$: $x(t) = \frac{2\pi B}{T^2} \sum_{k=-\infty}^{\infty} \left( \frac{1}{2a} e^{-a|kT|} + e^{-akT} u(kT) \right) \delta(t - kT)$

Esto representa un tren de impulsos delta de Dirac ponderados en el dominio del tiempo, localizados en múltiplos enteros del período $T$$T$. Los pesos están dados por los términos entre corchetes.

### Consulta sobre Modulación por Amplitud (AM) con Detección Coherente:

La modulación AM (Amplitud Modulada) es una técnica de modulación lineal en la que la amplitud de una portadora de alta frecuencia varía de acuerdo con la amplitud de la señal de mensaje. La expresión general de la señal modulada es:

$s(t) = A_c [1 + m(t)] \cos(2\pi f_c t)$

Donde:

- $A_c$: amplitud de la señal portadora.
- $m(t)$: señal mensaje (normalizada para que $|m(t)| \le 1$ y evitar sobremodulación).
- $f_c$: frecuencia de la portadora.

### Detección Coherente (Sincrónica):

Este método de demodulación requiere una señal local que tenga la misma frecuencia y fase que la portadora original. El procedimiento consiste en multiplicar la señal AM recibida por una portadora local y luego aplicar un filtro pasa-bajos (LPF) para recuperar la información.

**Etapas del proceso:**

- Señal recibida: $s(t) = A_c [1 + m(t)] \cos(2\pi f_c t)$
- Portadora local: $c_{local}(t) = \cos(2\pi f_c t + \phi)$  
  donde $\phi$ es el desfase entre la portadora recibida y la local. Si $\phi = 0$, la detección es ideal.

**Multiplicación de señales:**

$r(t) = s(t) \cdot c_{local}(t) = A_c [1 + m(t)] \cos(2\pi f_c t) \cos(2\pi f_c t + \phi)$

Aplicando la identidad trigonométrica:
$\cos A \cos B = \frac{1}{2}[\cos(A - B) + \cos(A + B)]$

$r(t) = \frac{A_c}{2} [1 + m(t)] \left[ \cos(\phi) + \cos(4\pi f_c t + \phi) \right]$

**Filtrado pasa-bajos (LPF):**

El LPF elimina la componente de alta frecuencia:
$y(t) = LPF\{r(t)\} = \frac{A_c}{2} \cos(\phi) [1 + m(t)]$

**En el caso ideal ($\phi = 0$):**
$y(t) = \frac{A_c}{2} [1 + m(t)] = \frac{A_c}{2} + \frac{A_c}{2} m(t)$

El primer término es una componente DC, que puede eliminarse con un filtro bloqueador de DC. La parte variable es proporcional a la señal de mensaje original.

### Aplicaciones de la Detección Coherente:

- **Comunicaciones:** mejor recuperación en ambientes ruidosos comparado con la detección por envolvente.
- **Receptores de alta fidelidad:** mejora la calidad del audio.
- **Modulaciones avanzadas:** fundamental para SSB (banda lateral única) y VSB (banda lateral vestigial).
- **Radionavegación:** se usa en algunos sistemas que emplean AM.

### Ventajas:

- Mayor resistencia al ruido.
- Compatible con SSB y VSB.
- No necesita portadora residual.

### Desventajas:

- Mayor complejidad del receptor (requiere sincronización de fase/frecuencia).
- Sensible a errores de fase.

### Ejemplo en Python:

A continuación, un ejemplo para simular la modulación AM y su detección coherente con señales tipo pulso y coseno usando `scipy.fft.rfft` para el análisis espectral.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
from scipy.signal import butter, lfilter

# --- Parámetros ---
Fs = 10000  # Frecuencia de muestreo
T = 1/Fs    # Período de muestreo
tiempo = 1  # Duración de la simulación en segundos
t = np.arange(0, tiempo, T) # Vector de tiempo

Fc = 1000 # Frecuencia de la portadora (debe ser mucho mayor que la frecuencia del mensaje)

# Índice de modulación (mu)
mu = 0.8

# --- Diseño del Filtro Pasa-Bajos (para demodulación) ---
orden = 5
frecuencia_corte = 2 * Fc / Fs # Frecuencia de corte normalizada
b, a = butter(orden, frecuencia_corte, btype='low')

# --- Función para Modulación AM ---
def modular_am(mensaje, Ac, Fc, Fs, mu):
    t = np.arange(0, len(mensaje)/Fs, 1/Fs)
    portadora = Ac * np.cos(2 * np.pi * Fc * t)
    senal_modulada = Ac * (1 + mu * mensaje) * np.cos(2 * np.pi * Fc * t)
    return senal_modulada, portadora

# --- Función para Demodulación Coherente ---
def demodular_am_coherente(senal_recibida, Fc, Fs, b, a):
    t = np.arange(0, len(senal_recibida)/Fs, 1/Fs)
    portadora_local = 2 * np.cos(2 * np.pi * Fc * t) # Se multiplica por 2 para compensar el factor 1/2 en la demodulación ideal
    senal_multiplicada = senal_recibida * portadora_local
    senal_demodulada_filtrada = lfilter(b, a, senal_multiplicada)
    # Eliminar componente DC para recuperar el mensaje (aproximado)
    senal_demodulada = (senal_demodulada_filtrada / Ac) - 1
    return senal_demodulada

# --- Parámetros de la señal mensaje ---
Am = 0.5 # Amplitud de la señal mensaje
Ac = 1   # Amplitud de la portadora

# --- Caso 1: Señal Mensaje tipo Pulso Rectangular ---
duracion_pulso = 0.1
pulso = np.zeros_like(t)
pulso[int(0.4*Fs):int(0.6*Fs)] = Am
mensaje_pulso = pulso

# Modulación
senal_modulada_pulso, portadora_pulso = modular_am(mensaje_pulso, Ac, Fc, Fs, mu)

# Demodulación
senal_demodulada_pulso = demodular_am_coherente(senal_modulada_pulso, Fc, Fs, b, a)

# --- Gráficas en el Dominio del Tiempo (Pulso Rectangular) ---
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, mensaje_pulso)
plt.title('Señal Mensaje (Pulso Rectangular) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0.3, 0.7]) # Enfocar en la región del pulso

plt.subplot(3, 1, 2)
plt.plot(t, senal_modulada_pulso)
plt.title('Señal Modulada AM (Pulso Rectangular) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0.3, 0.7]) # Enfocar en la región del pulso

plt.subplot(3, 1, 3)
plt.plot(t, senal_demodulada_pulso)
plt.title('Señal Demodulada AM Coherente (Pulso Rectangular) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0.3, 0.7]) # Enfocar en la región del pulso

plt.tight_layout()
plt.show()

# --- Gráficas en el Dominio de la Frecuencia (Pulso Rectangular) ---
# Calcular la transformada de Fourier real
mensaje_pulso_fft = rfft(mensaje_pulso)
senal_modulada_pulso_fft = rfft(senal_modulada_pulso)
senal_demodulada_pulso_fft = rfft(senal_demodulada_pulso)

# Obtener las frecuencias correspondientes
frecuencias = rfftfreq(len(t), 1/Fs)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(frecuencias, np.abs(mensaje_pulso_fft))
plt.title('Espectro de la Señal Mensaje (Pulso Rectangular)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de interés

plt.subplot(3, 1, 2)
plt.plot(frecuencias, np.abs(senal_modulada_pulso_fft))
plt.title('Espectro de la Señal Modulada AM (Pulso Rectangular)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2 + 200]) # Rango de frecuencias de interés

plt.subplot(3, 1, 3)
plt.plot(frecuencias, np.abs(senal_demodulada_pulso_fft))
plt.title('Espectro de la Señal Demodulada AM Coherente (Pulso Rectangular)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de interés

plt.tight_layout()
plt.show()

# --- Caso 2: Señal Mensaje tipo Coseno ---
Fm = 10  # Frecuencia del mensaje
mensaje_coseno = Am * np.cos(2 * np.pi * Fm * t)

# Modulación
senal_modulada_coseno, portadora_coseno = modular_am(mensaje_coseno, Ac, Fc, Fs, mu)

# Demodulación
senal_demodulada_coseno = demodular_am_coherente(senal_modulada_coseno, Fc, Fs, b, a)

# --- Gráficas en el Dominio del Tiempo (Coseno) ---
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, mensaje_coseno)
plt.title('Señal Mensaje (Coseno) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0, 0.5]) # Mostrar varias oscilaciones del mensaje

plt.subplot(3, 1, 2)
plt.plot(t, senal_modulada_coseno)
plt.title('Señal Modulada AM (Coseno) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0, 0.5]) # Mostrar varias oscilaciones del mensaje

plt.subplot(3, 1, 3)
plt.plot(t, senal_demodulada_coseno)
plt.title('Señal Demodulada AM Coherente (Coseno) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0, 0.5]) # Mostrar varias oscilaciones del mensaje

plt.tight_layout()
plt.show()

# --- Gráficas en el Dominio de la Frecuencia (Coseno) ---
# Calcular la transformada de Fourier real
mensaje_coseno_fft = rfft(mensaje_coseno)
senal_modulada_coseno_fft = rfft(senal_modulada_coseno)
senal_demodulada_coseno_fft = rfft(senal_demodulada_coseno)

# Obtener las frecuencias correspondientes
frecuencias = rfftfreq(len(t), 1/Fs)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(frecuencias, np.abs(mensaje_coseno_fft))
plt.title('Espectro de la Señal Mensaje (Coseno)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de interés

plt.subplot(3, 1, 2)
plt.plot(frecuencias, np.abs(senal_modulada_coseno_fft))
plt.title('Espectro de la Señal Modulada AM (Coseno)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2 + 200]) # Rango de frecuencias de interés

plt.subplot(3, 1, 3)
plt.plot(frecuencias, np.abs(senal_demodulada_coseno_fft))
plt.title('Espectro de la Señal Demodulada AM Coherente (Coseno)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de interés

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import rfft, irfft, rfftfreq
from IPython.display import Audio, display

# --- 1. Cargar audio ---
fs, audio = wavfile.read("fragmento_cancion.wav")

# Convertir a mono si es estéreo
if audio.ndim == 2:
    audio = audio[:, 0]

# Normalizar
audio = audio / np.max(np.abs(audio))

# Fragmento a usar (e.g., 5 segundos)
dur = 5
N = int(fs * dur)
mensaje = audio[:N]
t = np.linspace(0, dur, N, endpoint=False)

# --- 2. Modulación AM ---
fc = 10000  # Frecuencia portadora
Ac = 1
m_index = 1

portadora = Ac * np.cos(2 * np.pi * fc * t)
s_am = Ac * (1 + m_index * mensaje) * portadora

# --- 3. Demodulación coherente (θ₀ = 0) ---
demod = s_am * portadora  # Mezcla coherente

# --- 4. Filtrado pasa bajas ideal (FFT) ---
Demod_F = rfft(demod)
f = rfftfreq(N, 1/fs)
cutoff = 5000  # Hz

# Filtro ideal: anula frecuencias mayores al corte
Demod_F_filtered = Demod_F.copy()
Demod_F_filtered[f > cutoff] = 0

mensaje_recuperado = irfft(Demod_F_filtered)

# --- 5. Función de gráfica ---
def plot_signal_and_spectrum(signal, title, fs):
    N_local = len(signal)
    t_local = np.linspace(0, N_local/fs, N_local, endpoint=False)
    f_local = rfftfreq(N_local, 1/fs)
    S = np.abs(rfft(signal))

    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(t_local, signal)
    plt.title(f'{title} - Tiempo')
    plt.xlabel('Tiempo [s]')

    plt.subplot(1, 2, 2)
    plt.plot(f_local, S)
    plt.title(f'{title} - Frecuencia')
    plt.xlabel('Frecuencia [Hz]')
    plt.tight_layout()
    plt.show()

# --- 6. Etapas del sistema ---

# Etapa 1: Mensaje original
plot_signal_and_spectrum(mensaje, "1. Mensaje original", fs)
print("🔊 Reproduciendo mensaje original:")
display(Audio(mensaje, rate=fs))

# Etapa 2: Señal AM
plot_signal_and_spectrum(s_am, "2. Señal AM (modulada)", fs)
print("🔊 Reproduciendo señal AM:")
display(Audio(s_am, rate=fs))

# Etapa 3: Mezcla coherente
plot_signal_and_spectrum(demod, "3. Señal demodulada (sin filtro)", fs)
print("🔊 Reproduciendo señal demodulada (sin filtro):")
display(Audio(demod, rate=fs))

# Etapa 4: Mensaje recuperado (filtrado)
plot_signal_and_spectrum(mensaje_recuperado, "4. Mensaje recuperado (filtrado)", fs)
print("🔊 Reproduciendo mensaje recuperado:")
display(Audio(mensaje_recuperado, rate=fs))

# Etapa 5: Comparación final
min_len = min(len(mensaje), len(mensaje_recuperado))
t_cmp = np.linspace(0, min_len/fs, min_len, endpoint=False)

plt.figure(figsize=(10, 4))
plt.plot(t_cmp, mensaje[:min_len], label='Mensaje original')
plt.plot(t_cmp, mensaje_recuperado[:min_len], label='Mensaje recuperado', linestyle='--')
plt.title('Comparación: mensaje original vs recuperado')
plt.xlabel('Tiempo [s]')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
import scipy.integrate as integrate

# --- Parámetros de simulación ---
f = 60  # Hz
Vp = 10 # V
fs_sim = 200 * f # Hz
duration_sim = 4 / f # segundos (4 ciclos)
N_sim = int(fs_sim * duration_sim)
t_sim = np.linspace(0, duration_sim, N_sim, endpoint=False)

st.subheader("Parámetros de Carga")
resistance = st.slider("Valor de la Resistencia R (Ohms)", 1.0, 100.0, 10.0, step=1.0)
capacitance_nf = st.slider("Valor del Condensador C (nF)", 0.0, 1000.0, 0.0, step=10.0)
capacitance = capacitance_nf * 1e-9 # Convertir nF a Faradios

# --- Cálculo de señales ---
v_source = Vp * np.sin(2 * np.pi * f * t_sim)
v_rectified = np.abs(v_source)

i_r_load = v_rectified / resistance

dt = 1.0 / fs_sim
i_rc_load = np.zeros(N_sim)
v_c = 0.0

if capacitance > 1e-12: # Solo si hay capacitor
    for k in range(N_sim - 1):
        i_rc_load[k] = (v_rectified[k] - v_c) / resistance
        dv_c = i_rc_load[k] * dt / capacitance
        v_c += dv_c
        # El capacitor no puede tener un voltaje mayor al pico de la fuente rectificada
        if v_c > v_rectified[k+1]:
             v_c = v_rectified[k+1]
else: # Comportamiento resistivo si C es muy pequeño
     i_rc_load = i_r_load


# --- Cálculo de FFT y THD ---

def calculate_thd(signal, fs, fundamental_freq, num_harmonics=10):
    """ Calcula el THD de una señal """
    N = len(signal)
    yf = fft(signal)
    xf = fftfreq(N, 1/fs)

    # Encontrar la fundamental
    # Buscar el pico cerca de la fundamental_freq (o 2*fundamental_freq para rectificada)
    # La frecuencia fundamental de la señal rectificada de 60Hz es 120Hz
    target_fund_freq = 2 * fundamental_freq
    fundamental_idx = np.argmin(np.abs(xf - target_fund_freq))
    fundamental_magnitude = 2.0/N * np.abs(yf[fundamental_idx])

    if fundamental_magnitude < 1e-9: # Evitar división por cero
        return np.nan, xf[0:N//2], 2.0/N * np.abs(yf[0:N//2]) # Retornar espectro unilateral

    harmonic_magnitudes_sq = 0
    max_harm_idx = 0
    freqs_unilateral = xf[0:N//2]
    magnitudes_unilateral = 2.0/N * np.abs(yf[0:N//2])

    for n in range(2, num_harmonics + 2): # Armónicos de la fundamental rectificada (2*f, 3*f, 4*f, etc.)
        harmonic_freq = n * target_fund_freq
        harmonic_idx = np.argmin(np.abs(xf - harmonic_freq))
        if harmonic_idx < N // 2: # Asegurarse de estar en el rango unilateral
             harmonic_magnitude = 2.0/N * np.abs(yf[harmonic_idx])
             harmonic_magnitudes_sq += harmonic_magnitude**2
             max_harm_idx = max(max_harm_idx, harmonic_idx)


    thd = np.sqrt(harmonic_magnitudes_sq) / fundamental_magnitude * 100

    # Recortar espectro para visualización hasta el último armónico considerado + un poco
    view_limit_hz = (num_harmonics + 2) * target_fund_freq * 1.2
    view_idx_limit = np.argmin(np.abs(freqs_unilateral - view_limit_hz)) + 10
    view_idx_limit = min(view_idx_limit, N // 2) # No exceder el límite de N//2

    return thd, freqs_unilateral[:view_idx_limit], magnitudes_unilateral[:view_idx_limit]


thd_r, freqs_r, mags_r = calculate_thd(i_r_load, fs_sim, f)
thd_rc, freqs_rc, mags_rc = calculate_thd(i_rc_load, fs_sim, f)

# Cálculo del Factor de Potencia por Distorsión
fp_distortion_r = 1 / np.sqrt(1 + (thd_r/100)**2) if not np.isnan(thd_r) else np.nan
fp_distortion_rc = 1 / np.sqrt(1 + (thd_rc/100)**2) if not np.isnan(thd_rc) else np.nan


# --- Mostrar resultados ---
st.subheader("Resultados de la Simulación")

col1, col2 = st.columns(2)

with col1:
    st.markdown("### Carga Resistiva (R)")
    st.write(f"Resistencia: {resistance} Ohms")
    st.write(f"THD de la Corriente: {thd_r:.2f} %" if not np.isnan(thd_r) else "THD no calculado")
    st.write(f"Factor de Potencia por Distorsión: {fp_distortion_r:.4f}" if not np.isnan(fp_distortion_r) else "FP Distorsión no calculado")


with col2:
    st.markdown("### Carga RC en Serie")
    st.write(f"Resistencia: {resistance} Ohms")
    st.write(f"Capacitancia: {capacitance_nf} nF")
    st.write(f"THD de la Corriente: {thd_rc:.2f} %" if not np.isnan(thd_rc) else "THD no calculado")
    st.write(f"Factor de Potencia por Distorsión: {fp_distortion_rc:.4f}" if not np.isnan(fp_distortion_rc) else "FP Distorsión no calculado")

# --- Graficar formas de onda ---
st.subheader("Formas de Onda de Corriente")
fig_waveforms, ax_waveforms = plt.subplots(figsize=(12, 6))
ax_waveforms.plot(t_sim, i_r_load, label="Carga Resistiva (R)", alpha=0.7)
if capacitance_nf > 0:
    ax_waveforms.plot(t_sim, i_rc_load, label=f"Carga RC (R={resistance}Ω, C={capacitance_nf}nF)", alpha=0.7)
ax_waveforms.set_xlabel("Tiempo [s]")
ax_waveforms.set_ylabel("Corriente [A]")
ax_waveforms.set_title("Corriente a través de la Carga")
ax_waveforms.legend()
ax_waveforms.grid(True)
st.pyplot(fig_waveforms)

# --- Graficar espectros de frecuencia ---
st.subheader("Espectros de Frecuencia de Corriente")
fig_spectrums, ax_spectrums = plt.subplots(figsize=(12, 6))
ax_spectrums.plot(freqs_r, mags_r, label="Espectro Corriente Carga R", alpha=0.7)
if capacitance_nf > 0:
    ax_spectrums.plot(freqs_rc, mags_rc, label=f"Espectro Corriente Carga RC (R={resistance}Ω, C={capacitance_nf}nF)", alpha=0.7)
ax_spectrums.set_xlabel("Frecuencia [Hz]")
ax_spectrums.set_ylabel("Magnitud")
ax_spectrums.set_title("Espectro de Frecuencia de la Corriente")
ax_spectrums.set_xlim(0, 15 * f) # Limitar eje X para mejor visualización de armónicos
ax_spectrums.legend()
ax_spectrums.grid(True)
st.pyplot(fig_spectrums)
```

```python
%%writefile 1_AM_Modulacion_🔉.py
import streamlit as st
import numpy as np
from scipy import signal
from scipy.io import wavfile
import matplotlib.pyplot as plt
import yt_dlp
import os

# --- Función personalizada para descarga y preprocesamiento de audio ---
def obtener_audio_de_youtube(enlace, tiempo_ini=20, tiempo_fin=25, frecuencia_objetivo=22050):
    """
    Extrae el audio de un video de YouTube, lo recorta al segmento deseado de 5 segundos, lo convierte a mono en caso de ser necesario, ajusta su frecuencia de muestreo y normaliza su nivel de volumen.

    """
    archivo_temporal = "temp_clip"
    opciones_yt = {
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'wav',
            'preferredquality': '192',
        }],
        'outtmpl': archivo_temporal,
        'quiet': True,
    }

    try:
        with yt_dlp.YoutubeDL(opciones_yt) as ydl:
            ydl.download([enlace])

        tasa_original, datos_audio = wavfile.read(archivo_temporal + ".wav")

        st.info(f"Usando el audio entre los segundos {tiempo_ini} y {tiempo_fin}.")
        i_inicio = int(tiempo_ini * tasa_original)
        i_final = int(tiempo_fin * tasa_original)

        if i_final > len(datos_audio):
            st.warning("El video tiene menos duración de la esperada. Se tomará el fragmento disponible.")
            i_final = len(datos_audio)
        if i_inicio >= len(datos_audio):
            st.error("El punto inicial está fuera del rango del audio.")
            os.remove(archivo_temporal + ".wav")
            return None, None

        fragmento = datos_audio[i_inicio:i_final]

        if fragmento.ndim > 1:
            fragmento = np.mean(fragmento, axis=1)

        if tasa_original != frecuencia_objetivo:
            cantidad_muestras = int(len(fragmento) * frecuencia_objetivo / tasa_original)
            fragmento = signal.resample(fragmento, cantidad_muestras)
            tasa_original = frecuencia_objetivo

        if np.max(np.abs(fragmento)) > 0:
            audio_final = fragmento / np.max(np.abs(fragmento))
        else:
            audio_final = fragmento

        os.remove(archivo_temporal + ".wav")
        return audio_final, tasa_original

    except Exception as error:
        st.error(f"Ocurrió un error al trabajar con el audio: {error}")
        return None, None

# --- Interfaz principal ---
st.set_page_config(layout="wide")
st.title("📻 Simulación de Modulación AM y su Demodulación")

st.markdown("""
Este módulo ilustra cómo una señal de audio puede ser transformada mediante modulación en amplitud (AM) y posteriormente reconstruida utilizando un detector coherente.
""")

st.markdown("""
La **modulación AM** facilita la transmisión de una señal de información al modificar la amplitud de una señal portadora. En esta aplicación, la señal del mensaje se obtiene a partir del audio de un video de YouTube.
""")

# --- Controles interactivos ---
st.sidebar.header("Configuración")
enlace_youtube = st.sidebar.text_input("Enlace del video:", "https://www.youtube.com/watch?v=dQw4w9WgXcQ")
amplitud_portadora = st.sidebar.slider("Amplitud (Ac)", 0.1, 5.0, 1.0, 0.1)
frecuencia_portadora = st.sidebar.slider("Frecuencia Portadora (Hz)", 500, 5000, 5000, 100)
indice_modulacion = st.sidebar.slider("Índice de Modulación (ka)", 0.1, 1.5, 1.0, 0.1)

if st.sidebar.button("Iniciar Modulación"):
    if not enlace_youtube:
        st.warning("Ingresa un enlace válido.")
    else:
        with st.spinner("Procesando el audio..."):
            mensaje, tasa = obtener_audio_de_youtube(enlace_youtube)

        if mensaje is not None:
            tiempo = np.linspace(0, len(mensaje) / tasa, len(mensaje), endpoint=False)

            st.header("1. Señal de Entrada")
            st.markdown("Este es el fragmento original que será utilizado como mensaje.")

            fig1, ax1 = plt.subplots(figsize=(12, 4))
            ax1.plot(tiempo, mensaje, color='steelblue')
            ax1.set_title("Mensaje de Audio Original")
            ax1.set_xlabel("Tiempo (s)")
            ax1.set_ylabel("Amplitud")
            ax1.grid(True)
            st.pyplot(fig1)
            st.audio(mensaje, sample_rate=tasa)

            st.header("2. Señal Modulada AM")
            st.markdown("Aplicamos la fórmula: $s(t) = A_c [1 + k_a m(t)] \\cos(2\\pi f_c t)$")

            with st.expander("Mostrar análisis espectral de la señal AM"):
                st.markdown(r"""
                Transformando $s(t)$ al dominio de la frecuencia obtenemos impulsos en ±fc más la señal mensaje desplazada.
                Es decir:
                $$ S(f) = \frac{A_c}{2}[\delta(f - f_c) + \delta(f + f_c)] + \frac{1}{2}[M(f - f_c) + M(f + f_c)] $$
                """)

            portadora = amplitud_portadora * np.cos(2 * np.pi * frecuencia_portadora * tiempo)
            señal_am = (1 + indice_modulacion * mensaje) * portadora

            fig2, ax2 = plt.subplots(figsize=(12, 4))
            ax2.plot(tiempo, señal_am, color='firebrick')
            ax2.set_title("Señal con Modulación AM")
            ax2.set_xlabel("Tiempo (s)")
            ax2.set_ylabel("Amplitud")
            ax2.grid(True)
            st.pyplot(fig2)
            st.audio(señal_am, sample_rate=tasa)

            st.header("3. Recuperación de la Señal (Demodulación)")
            st.subheader("Paso 3.1: Multiplicación por Portadora")
            st.markdown("Se utiliza la misma portadora para llevar de nuevo la señal al centro del espectro.")

            señal_multiplicada = señal_am * portadora
            fig3, ax3 = plt.subplots(figsize=(12, 4))
            ax3.plot(tiempo, señal_multiplicada, color='darkorange')
            ax3.set_title("Producto de la Señal AM con la Portadora")
            ax3.set_xlabel("Tiempo (s)")
            ax3.set_ylabel("Amplitud")
            ax3.grid(True)
            st.pyplot(fig3)
            st.audio(señal_multiplicada, sample_rate=tasa)
            st.warning("El sonido es ruidoso debido a la mezcla de frecuencias.")

            st.subheader("Paso 3.2: Filtrado Paso Bajo")
            frecuencia_corte = 4000
            st.markdown(f"Usamos un filtro que solo deja pasar componentes hasta {frecuencia_corte} Hz.")

            orden = 4
            wn = 2 * frecuencia_corte / tasa
            b, a = signal.butter(orden, wn, btype='low')
            señal_filtrada = signal.lfilter(b, a, señal_multiplicada)

            fig4, ax4 = plt.subplots(figsize=(12, 4))
            ax4.plot(tiempo, señal_filtrada, color='seagreen')
            ax4.set_title("Señal Filtrada")
            ax4.set_xlabel("Tiempo (s)")
            ax4.set_ylabel("Amplitud")
            ax4.grid(True)
            st.pyplot(fig4)
            st.audio(señal_filtrada, sample_rate=tasa)

            st.subheader("Paso 3.3: Ajuste de Amplitud")
            st.markdown("La señal se amplifica para recuperar su escala original.")
            ganancia_recuperacion = 2 / (amplitud_portadora ** 2)
            señal_final = señal_filtrada * ganancia_recuperacion

            fig5, ax5 = plt.subplots(figsize=(12, 4))
            ax5.plot(tiempo, señal_final, color='dodgerblue', label='Recuperada')
            ax5.plot(tiempo, mensaje, 'r--', label='Original', alpha=0.6)
            ax5.set_title("Comparación de Señales: Original vs Recuperada")
            ax5.set_xlabel("Tiempo (s)")
            ax5.set_ylabel("Amplitud")
            ax5.legend()
            ax5.grid(True)
            st.pyplot(fig5)
            st.audio(señal_final, sample_rate=tasa)
            st.success("¡Demodulación completada exitosamente!")
```

```python
%%writefile 2_Factor_de_Potencia_THD_⚡.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
import scipy.integrate as integrate

st.set_page_config(
    page_title="Factor de Potencia (THD)",
    page_icon="⚡",
    layout="wide",
)

st.markdown(r"""
# Factor de Potencia y Distorsión Armónica Total (THD) ⚡

### ¿Qué representan la THD y el Factor de Potencia?

En los sistemas eléctricos, las formas de onda de corriente o voltaje con frecuencia se desvían de una sinusoide ideal, especialmente cuando se conectan dispositivos no lineales como variadores de velocidad o fuentes conmutadas. Estas alteraciones se deben a la aparición de armónicos, que son componentes con frecuencias múltiplos de la frecuencia fundamental.

- **Distorsión Armónica Total (THD):** Representa una métrica que cuantifica cuánto se ha distorsionado una señal con respecto a una onda sinusoidal pura. Se calcula como la relación entre la raíz cuadrada de la suma de los cuadrados de todas las componentes armónicas y el valor de la fundamental:

$$
THD_I = \frac{\sqrt{I_2^2 + I_3^2 + I_4^2 + \dots}}{I_1} \times 100\%
$$

$$
THD_V = \frac{\sqrt{V_2^2 + V_3^2 + V_4^2 + \dots}}{V_1} \times 100\%
$$

Donde $I_n$ y $V_n$ son los valores RMS de los armónicos, y $I_1$, $V_1$ son los valores RMS de la fundamental.

- **Factor de Potencia (FP):** Mide cuán eficientemente se convierte la potencia suministrada en trabajo útil. Se define como la razón entre la potencia activa y la potencia aparente, y se expresa como el producto de dos factores:

$$
FP = FP_{desplazamiento} \cdot FP_{distorsión}
$$

  - **Por desplazamiento ($\cos(\phi)$):** Hace referencia al ángulo de desfase entre las componentes fundamentales del voltaje y la corriente.
  - **Por distorsión:** Considera la pérdida de eficiencia causada por armónicos en la señal.

Cuando hay distorsión en la señal, el FP total disminuye.

### Cálculo del THD con FFT

La Transformada Rápida de Fourier (FFT) permite descomponer una señal en sus frecuencias constituyentes, identificando tanto la fundamental como sus armónicos.

Para una señal $x(t)$ discretizada, la FFT nos permite:

1. Calcular la FFT: `X = fft(x)`
2. Obtener frecuencias: `freqs = fftfreq(N, 1/fs)`
3. Identificar el índice $k_1$ correspondiente a la frecuencia fundamental.
4. Hallar los índices $k_n$ de los armónicos.
5. Calcular RMS de la fundamental: $I_1 = |X[k_1]| \cdot \sqrt{2} / N$
6. Calcular RMS de los armónicos: $I_n = |X[k_n]| \cdot \sqrt{2} / N$
7. Aplicar la fórmula de THD.

Si se usa un espectro unilateral ($P1[k]$):

$$
THD_I = \frac{\sqrt{\sum_{n=2}^{N_{harm}} P1[k_n]^2}}{P1[k_1]}
$$

### Cómo afecta la THD al Factor de Potencia

El factor de potencia total, considerando la distorsión, se calcula como:

$$
FP = \frac{P}{S} = \frac{\text{Potencia Activa}}{\text{Potencia Aparente}}
$$

La potencia aparente se determina con los valores RMS totales del voltaje y la corriente, que incluyen armónicos:

$$
X_{RMS} = X_1 \sqrt{1 + THD_X^2}
$$

La potencia activa se suele aproximar con la fundamental:

$$
P = V_1 I_1 \cos(\phi_1)
$$

Entonces, el componente de distorsión del factor de potencia se calcula como:

$$
FP_{distorsión} = \frac{I_1}{I_{RMS}} = \frac{1}{\sqrt{1 + THD_I^2}}
$$

En situaciones donde el voltaje tiene muy poca distorsión:

$$
FP = \cos(\phi_1) \cdot \frac{1}{\sqrt{1 + THD_I^2}}
$$

Si $\phi_1 = 0$ (carga puramente resistiva), entonces $\cos(\phi_1) = 1$ y el FP depende únicamente de la THD de la corriente.

### Ejemplo: Rectificador de Onda Completa

Se analiza una carga alimentada por un rectificador de onda completa con una entrada sinusoidal, en dos configuraciones: solo con resistencia (R), y con resistencia en serie con un condensador (RC).

Para $v(t) = V_p \sin(2\pi f t)$, la salida del rectificador será $|v(t)| = V_p |\sin(2\pi f t)|$.

#### i) Carga Resistiva

La corriente será: $i(t) = |v(t)| / R = (V_p/R) |\sin(2\pi f t)|$

#### ii) Carga RC en Serie

En este caso, la corriente depende de cómo actúa el condensador. La corriente solo fluye cuando el voltaje del rectificador supera al del capacitor, lo que da lugar a picos de corriente muy estrechos y con alta distorsión.

### Simulación

Usaremos:

- Frecuencia: $f = 60$ Hz
- Voltaje pico: $V_p = 10$ V
- Frecuencia de muestreo: $fs = 200 \cdot f$
- Duración: varios ciclos

Se incluyen controles para ajustar R y C:
""")

# --- Parámetros de la simulación ---
f = 60  # Frecuencia de la red en Hz
Vp = 10 # Voltaje pico en V
fs_sim = 200 * f  # Frecuencia de muestreo

duration_sim = 4 / f  # Tiempo total: 4 ciclos
N_sim = int(fs_sim * duration_sim)  # Número total de muestras
t_sim = np.linspace(0, duration_sim, N_sim, endpoint=False)  # Vector de tiempo

# Interfaz de usuario para parámetros de carga
st.subheader("Parámetros de la Carga")
resistance = st.slider("Resistencia R (Ohms)", 1.0, 100.0, 10.0, step=1.0)
capacitance_nf = st.slider("Capacitancia C (nF)", 0.0, 1000.0, 0.0, step=10.0)
capacitance = capacitance_nf * 1e-9  # Conversión a Faradios

# --- Generación de señales ---
v_source = Vp * np.sin(2 * np.pi * f * t_sim)  # Fuente de voltaje
v_rectified = np.abs(v_source)  # Voltaje rectificado
i_r_load = v_rectified / resistance  # Corriente para carga resistiva

# Simulación de la respuesta de una carga RC
# Se usa un método de integración numérica (Euler) para resolver:
# dv_c/dt = (v_rectified - v_c)/(R*C)
dt = 1.0 / fs_sim
i_rc_load = np.zeros(N_sim)
v_c = 0.0  # Voltaje inicial en el capacitor

if capacitance > 1e-12:  # Solo simula si C es relevante
    for k in range(N_sim - 1):
        i_rc_load[k] = (v_rectified[k] - v_c) / resistance
        dv_c = i_rc_load[k] * dt / capacitance
        v_c += dv_c
        if v_c > v_rectified[k+1]:
            v_c = v_rectified[k+1]  # Límite superior natural del capacitor
else:
    i_rc_load = i_r_load  # Si no hay capacitor, se comporta como carga R pura

# --- Cálculo de THD mediante FFT ---
def calculate_thd(signal, fs, fundamental_freq, num_harmonics=10):
    N = len(signal)
    yf = fft(signal)
    xf = fftfreq(N, 1/fs)

    # Se analiza la frecuencia fundamental de la señal rectificada (2*f)
    target_fund_freq = 2 * fundamental_freq
    fundamental_idx = np.argmin(np.abs(xf - target_fund_freq))
    fundamental_magnitude = 2.0/N * np.abs(yf[fundamental_idx])

    if fundamental_magnitude < 1e-9:
        return np.nan, xf[0:N//2], 2.0/N * np.abs(yf[0:N//2])

    harmonic_magnitudes_sq = 0
    freqs_unilateral = xf[0:N//2]
    magnitudes_unilateral = 2.0/N * np.abs(yf[0:N//2])

    for n in range(2, num_harmonics + 2):
        harmonic_freq = n * target_fund_freq
        harmonic_idx = np.argmin(np.abs(xf - harmonic_freq))
        if harmonic_idx < N // 2:
            harmonic_magnitude = 2.0/N * np.abs(yf[harmonic_idx])
            harmonic_magnitudes_sq += harmonic_magnitude**2

    thd = np.sqrt(harmonic_magnitudes_sq) / fundamental_magnitude * 100

    view_limit_hz = (num_harmonics + 2) * target_fund_freq * 1.2
    view_idx_limit = np.argmin(np.abs(freqs_unilateral - view_limit_hz)) + 10
    view_idx_limit = min(view_idx_limit, N // 2)

    return thd, freqs_unilateral[:view_idx_limit], magnitudes_unilateral[:view_idx_limit]

# Aplicar la función para ambas cargas
thd_r, freqs_r, mags_r = calculate_thd(i_r_load, fs_sim, f)
thd_rc, freqs_rc, mags_rc = calculate_thd(i_rc_load, fs_sim, f)

# Calcular el Factor de Potencia por Distorsión
fp_distortion_r = 1 / np.sqrt(1 + (thd_r/100)**2) if not np.isnan(thd_r) else np.nan
fp_distortion_rc = 1 / np.sqrt(1 + (thd_rc/100)**2) if not np.isnan(thd_rc) else np.nan

# --- Mostrar resultados ---
st.subheader("Resultados de la Simulación")
col1, col2 = st.columns(2)

with col1:
    st.markdown("### Carga R únicamente")
    st.write(f"Resistencia: {resistance} Ohms")
    st.write(f"THD de Corriente: {thd_r:.2f} %" if not np.isnan(thd_r) else "THD no disponible")
    st.write(f"Factor de Potencia (Distorsión): {fp_distortion_r:.4f}" if not np.isnan(fp_distortion_r) else "FP no disponible")

with col2:
    st.markdown("### Carga RC combinada")
    st.write(f"Resistencia: {resistance} Ohms")
    st.write(f"Capacitancia: {capacitance_nf} nF")
    st.write(f"THD de Corriente: {thd_rc:.2f} %" if not np.isnan(thd_rc) else "THD no disponible")
    st.write(f"Factor de Potencia (Distorsión): {fp_distortion_rc:.4f}" if not np.isnan(fp_distortion_rc) else "FP no disponible")

# --- Gráfica de señales en el tiempo ---
st.subheader("Formas de Onda de la Corriente")
fig_waveforms, ax_waveforms = plt.subplots(figsize=(12, 6))
ax_waveforms.plot(t_sim, i_r_load, label="Carga R", alpha=0.7)
if capacitance_nf > 0:
    ax_waveforms.plot(t_sim, i_rc_load, label=f"Carga RC (R={resistance}Ω, C={capacitance_nf}nF)", alpha=0.7)
ax_waveforms.set_xlabel("Tiempo [s]")
ax_waveforms.set_ylabel("Corriente [A]")
ax_waveforms.set_title("Corriente entregada a la carga")
ax_waveforms.legend()
ax_waveforms.grid(True)
st.pyplot(fig_waveforms)

# --- Gráfica de espectros de frecuencia ---
st.subheader("Espectros de Frecuencia de Corriente")
fig_spectrums, ax_spectrums = plt.subplots(figsize=(12, 6))
ax_spectrums.plot(freqs_r, mags_r, label="Carga R", alpha=0.7)
if capacitance_nf > 0:
    ax_spectrums.plot(freqs_rc, mags_rc, label=f"Carga RC (R={resistance}Ω, C={capacitance_nf}nF)", alpha=0.7)
ax_spectrums.set_xlabel("Frecuencia [Hz]")
ax_spectrums.set_ylabel("Magnitud")
ax_spectrums.set_title("Contenido Armónico de la Corriente")
ax_spectrums.set_xlim(0, 15 * f)
ax_spectrums.legend()
ax_spectrums.grid(True)
st.pyplot(fig_spectrums)

# --- Discusión de Resultados ---
st.subheader("Discusión de Resultados")
st.markdown("""
**Carga Resistiva Pura (R):**
- La forma de la corriente replica la señal de voltaje rectificada: una onda sinusoidal cuyos semiciclos negativos han sido reflejados positivamente.

- En el análisis espectral, se identifica una frecuencia dominante en 120 Hz (el doble de la frecuencia de entrada), acompañada de armónicos pares como 240 Hz, 360 Hz, etc.

- DDebido a que la forma de la onda no es completamente senoidal, se presenta una distorsión armónica significativa.

- Aunque la corriente y el voltaje permanecen sincronizados en fase —lo que significa un factor de potencia por desplazamiento ideal (1), el factor de potencia global se reduce debido a los armónicos presentes.

**Carga RC (Resistencia y Condensador en Serie):**
- La incorporación del condensador modifica drásticamente la forma de la corriente, generando picos cortos y de gran amplitud cuando la tensión excede al voltaje acumulado en el condensador.

- Esto conlleva una corriente mucho más alterada en comparación con la carga puramente resistiva.

- En el dominio de la frecuencia, el espectro se amplía, apareciendo armónicos más intensos y componentes a frecuencias más altas.

- El contenido armónico (THD) de la corriente se eleva considerablemente por la influencia del condensador.

- Esta distorsión afecta negativamente al factor de potencia por distorsión, el cual disminuye conforme el THD aumenta, a pesar de que el desfase entre voltaje y corriente pueda seguir siendo bajo.

**Conclusiones Generales:**
- Es posible visualizar cómo influyen los valores de R y C usando los controles deslizantes incluidos en la simulación.

- A medida que se incrementa la capacitancia, los pulsos de corriente se vuelven más angostos y potentes, lo que intensifica la presencia de armónicos.

- Una corriente con alto THD reduce el factor de potencia total y puede provocar efectos adversos como sobrecalentamiento en transformadores, mayor carga en los conductores de neutro, entre otros.

- Es importante recordar que esta simulación es una aproximación; un análisis más riguroso requeriría modelar el circuito con componentes reales, incluyendo los diodos y las interacciones completas entre fuente, rectificador y carga.
""")
```

# **Detector de Genero **

```python
%%writefile 3_Clasificador_Genero_Musical.py
import streamlit as st
import yt_dlp
import os
import soundfile as sf
import numpy as np
import librosa
import joblib
import io

# --- Funciones Utilitarias --- #

def obtener_audio_desde_youtube(enlace, carpeta_destino='audio_descargado'):
    """
    Descarga el audio de un video de YouTube y lo convierte a formato WAV.
    """
    if not os.path.exists(carpeta_destino):
        os.makedirs(carpeta_destino)

    opciones_descarga = {
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'wav',
            'preferredquality': '192',
        }],
        'outtmpl': os.path.join(carpeta_destino, '%(title)s.%(ext)s'),
        'quiet': True,
        'noplaylist': True,
    }

    try:
        with yt_dlp.YoutubeDL(opciones_descarga) as ydl:
            informacion = ydl.extract_info(enlace, download=True)
            nombre_base = ydl.prepare_filename(informacion)
            nombre_final = os.path.splitext(nombre_base)[0] + '.wav'
            return nombre_final
    except Exception as error:
        st.error(f"No fue posible obtener el audio: {error}")
        return None

def procesar_audio_extraer_mfcc(ruta_audio, longitud_deseada):
    """
    Extrae MFCCs de un archivo de audio y los ajusta a una longitud fija.
    """
    try:
        señal, tasa_muestreo = librosa.load(ruta_audio, sr=None, duration=60)
        coef_mfcc = librosa.feature.mfcc(y=señal, sr=tasa_muestreo, n_mfcc=20)
        caracteristicas = coef_mfcc.flatten()

        if len(caracteristicas) > longitud_deseada:
            caracteristicas = caracteristicas[:longitud_deseada]
        else:
            caracteristicas = np.pad(caracteristicas, (0, longitud_deseada - len(caracteristicas)), mode='constant')

        return caracteristicas
    except Exception as error:
        st.error(f"Error al analizar el archivo de audio: {error}")
        return None

# --- Carga del Modelo de Clasificación --- #

@st.cache_resource
def cargar_modelo_desde_archivo(ruta):
    """
    Carga un modelo almacenado en un archivo .pkl.
    """
    try:
        with open(ruta, 'rb') as archivo:
            return joblib.load(archivo)
    except FileNotFoundError:
        return None

# --- Inicialización del Modelo por Defecto --- #

if 'modelo_datos' not in st.session_state:
    ruta_modelo_defecto = 'pop_vs_reggaeton.pkl'
    try:
        with st.spinner(f"Cargando el modelo predeterminado: {ruta_modelo_defecto}..."):
            datos_modelo = cargar_modelo_desde_archivo(ruta_modelo_defecto)

        if datos_modelo:
            st.session_state['modelo_datos'] = datos_modelo
            st.session_state['nombre_modelo'] = ruta_modelo_defecto
        else:
            st.session_state['modelo_datos'] = None
            st.session_state['nombre_modelo'] = None
    except EOFError:
        st.session_state['modelo_datos'] = None
        st.session_state['nombre_modelo'] = None
    except Exception as error:
        st.error(f"No se pudo cargar el modelo '{ruta_modelo_defecto}': {error}")
        st.session_state['modelo_datos'] = None
        st.session_state['nombre_modelo'] = None

# --- Interfaz Gráfica --- #

st.title("Clasificador de Género Musical 🎵")

opciones_menu = ["Inicio", "Subir Modelo", "Clasificar Género"]
seleccion = st.sidebar.selectbox("Menú Principal", opciones_menu)

if seleccion == "Inicio":
    st.header("Bienvenido/a")
    st.markdown("""
        Esta herramienta usa aprendizaje automático para identificar géneros musicales a partir de fragmentos de canciones.
    """)

    if st.session_state.get('nombre_modelo') is None:
        st.error("El modelo por defecto no se cargó correctamente. Por favor, verifica que el archivo exista o súbelo manualmente.")
    st.info(f"**Modelo en uso:** `{st.session_state.get('nombre_modelo', 'Ninguno')}`")

    st.markdown("""
        **¿Cómo usar esta app?**
        1. El sistema intentará cargar un modelo base llamado `pop_vs_reggaeton.pkl`.
        2. Si quieres usar otro, ve a la pestaña *Subir Modelo*.
        3. En la sección *Clasificar Género*, introduce un enlace de YouTube para clasificar el audio.
    """)

elif seleccion == "Subir Modelo":
    st.header("Cargar un Modelo de Clasificación")
    st.write("Puedes cargar un modelo en formato `.pkl` para reemplazar el predeterminado.")

    if st.session_state.get('nombre_modelo'):
        st.info(f"Modelo activo: **{st.session_state['nombre_modelo']}**")

    archivo_modelo = st.file_uploader("Selecciona un archivo .pkl", type=["pkl"])

    if archivo_modelo is not None:
        try:
            contenido = archivo_modelo.getvalue()
            datos_modelo = joblib.load(io.BytesIO(contenido))

            if isinstance(datos_modelo, dict) and 'modelo' in datos_modelo:
                st.session_state['modelo_datos'] = datos_modelo
                st.session_state['nombre_modelo'] = archivo_modelo.name
                st.success(f"Modelo '{archivo_modelo.name}' cargado correctamente.")
                st.info("Ahora puedes usarlo en la sección 'Clasificar Género'.")
            else:
                st.error("El archivo no tiene la estructura correcta. Se espera un diccionario con una clave 'modelo'.")

        except Exception as error:
            st.error(f"No fue posible cargar el modelo: {error}")

elif seleccion == "Clasificar Género":
    st.header("Identificación del Género Musical")

    if 'modelo_datos' not in st.session_state or st.session_state.get('modelo_datos') is None:
        st.warning("No se ha cargado ningún modelo. Usa el modelo por defecto o sube uno manualmente.")
        st.stop()

    datos_modelo = st.session_state['modelo_datos']
    clasificador = datos_modelo['modelo']
    num_caracteristicas = clasificador.n_features_in_

    st.info(f"Modelo activo: **{st.session_state['nombre_modelo']}**")

    enlace_youtube = st.text_input("Introduce el enlace de YouTube de la canción a analizar")

    if enlace_youtube:
        if st.button("Clasificar"):
            try:
                with st.spinner("Procesando el audio..."):
                    ruta_audio = obtener_audio_desde_youtube(enlace_youtube)

                    if ruta_audio:
                        caracteristicas = procesar_audio_extraer_mfcc(ruta_audio, longitud_deseada=num_caracteristicas)
                        os.remove(ruta_audio)

                        if caracteristicas is not None:
                            entrada = caracteristicas.reshape(1, -1)
                            prediccion = clasificador.predict(entrada)
                            probabilidad = clasificador.predict_proba(entrada)

                            st.markdown(f"### Género estimado: **{prediccion[0]}**")
                            st.markdown("---")
                            st.markdown("#### Detalle de probabilidades:")

                            for i, etiqueta in enumerate(clasificador.classes_):
                                st.write(f"- **{etiqueta}:** {probabilidad[0][i]*100:.2f}%")

            except Exception as error:
                st.error(f"Ocurrió un error durante la predicción: {error}")
```
